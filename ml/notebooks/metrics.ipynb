```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def product(a, b):
    return a * b


def arithmetic_mean(a, b):
    return (a + b) / 2


def harmonic_mean(a, b):
    return 2 * a * b / (a + b + 1e-6)


def geometric_mean(a, b):
    return np.sqrt(a * b)


def minimum(a, b):
    return np.minimum(a, b)


def maximum(a, b):
    return np.maximum(a, b)


def euclidean(a, b):
    return 1 - np.sqrt((a - 1) ** 2 + (b - 1) ** 2)


metrics = {
    r"Product $a \cdot b$": product,
    r"Arithmetic Mean $\frac{a + b}{2}$": arithmetic_mean,
    r"Harmonic Mean $\frac{2ab}{a + b}$": harmonic_mean,
    r"Geometric Mean $\sqrt{ab}$": geometric_mean,
    r"Minimum $\min(a, b)$": minimum,
    r"Maximum $\max(a, b)$": maximum,
    r"Euclidean $1 - \sqrt{(a - 1)^2 + (b - 1)^2}$": euclidean,
}

xx, yy = np.meshgrid(np.linspace(0, 1, 100), np.linspace(0, 1, 100))

for metric in metrics:
    fig, ax = plt.subplots(figsize=(12, 10))
    plt.title(metric)
    contour = plt.contourf(xx, yy, metrics[metric](xx, yy), levels=20, cmap="coolwarm")
    plt.xlabel("a")
    plt.ylabel("b")
    plt.colorbar(contour)
    ax.set_aspect("equal")
    plt.tight_layout()
    plt.show()
    plt.close()
```
